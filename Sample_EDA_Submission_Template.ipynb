## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of the dataset
df1 = df.copy()

# Remove duplicate records
df1 = df1.drop_duplicates()

# Fill missing values
df1["member_gender"] = df1["member_gender"].fillna("Unknown")
df1["member_birth_year"] = df1["member_birth_year"].fillna(df1["member_birth_year"].median())

# Convert date columns
df1["start_time"] = pd.to_datetime(df1["start_time"])
df1["end_time"] = pd.to_datetime(df1["end_time"])

# Create new features
df1["duration_min"] = df1["duration_sec"] / 60
df1["hour"] = df1["start_time"].dt.hour
df1["day"] = df1["start_time"].dt.day
df1["month"] = df1["start_time"].dt.month
df1["weekday"] = df1["start_time"].dt.day_name()
df1["age"] = 2018 - df1["member_birth_year"]

# Display first five rows
df1.head()
````

### What all manipulations have you done and insights you found?

The following data preprocessing steps were performed:

- Removed duplicate records.
- Filled missing values in member_gender and member_birth_year.
- Converted start_time and end_time into datetime format.
- Created new features such as duration_min, hour, day, month, weekday and age.
- Prepared the dataset for visualization and machine learning.

Insights:
The dataset became clean and analysis-ready. Feature engineering provided additional variables that help understand user riding behaviour and improve model performance.
